Q(7) What is the difference between “==”and “is” operators in python ?
- In Python, == and is are both comparison operators, but they serve different purposes:
== operator:
Checks for equality: This operator compares the values of two objects.
Returns True: if the values are equal.
Returns False: if the values are not equal.
Example:Python
x = [1, 2, 3]
y = [1, 2, 3]
print(x == y)  # Output: True (values are the same)
is operator:
Checks for identity: This operator compares the memory addresses of two objects.
Returns True: if both variables refer to the exact same object in memory.
Returns False: if the variables refer to different objects, even if their values are the same.
Example:Python
x = [1, 2, 3]
y = x
print(x is y)  # Output: True (both variables point to the same object)
z = [1, 2, 3]print(x is z)  # Output: False (different objects in memory)
In summary:
Use == to compare values.
Use is to check if two variables refer to the same object.
Use is when comparing with None. (e.g., if x is None.

Q(12) How does the elif  statement work ?
- In Python, elif stands for "else if". It's used in conjunction with if statements to create a chain of conditional checks.
How it works:
Initial if:
The first if statement is evaluated. If its condition is true, the code block associated with that if statement is executed, and the rest of the elif and else blocks are skipped.
elif checks:
If the initial if condition is false, the interpreter moves on to the first elif statement. If the condition in the elif statement is true, the code block associated with that elif statement is executed, and the rest of the elif and else blocks are skipped.
Multiple elif statements:
You can have multiple elif statements following an if statement. They are evaluated in order, and the first elif statement whose condition is true will have its code block executed.
else block (optional):
If none of the if or elif conditions are true, the code block associated with the optional else statement is executed.
Example:
Python
x = 10
if x > 15:
    print("x is greater than 15")
elif x > 5:
    print("x is greater than 5 but less than or equal to 15")
else:
    print("x is less than or equal to 5")
In this example, the output will be "x is greater than 5 but less than or equal to 15" because the first if condition is false, but the first elif condition is true.

(14)  Describe a scenario where a while loop is more suitable than a for loop ?
- A while loop is more suitable than a for loop in scenarios where the number of iterations is not predetermined and depends on a certain condition being met during the execution of the loop.
For example, consider a situation where you are writing a program to read user input until the user types "exit." In this case, you don't know how many times the user will input data, so a while loop would be ideal. Here’s how it could look in pseudocode:
    user_input = ""
    while user_input != "exit":
    user_input = getUserInput()  // Function to get input from the user
    processInput(user_input)      // Function to process the input
In this scenario, the loop continues to run as long as the user does not type "exit." The condition for continuing the loop is based on user input, which makes a while loop the most suitable choice. If you were to use a for loop here, you would have to set an arbitrary limit on the number of iterations, which wouldn't be practical or effective for this situation.
